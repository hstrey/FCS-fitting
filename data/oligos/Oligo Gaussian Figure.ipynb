```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
def makefig(datafile,outputfile):
    oligo=pd.read_csv(datafile)
    matplotlib.rcParams['figure.figsize'] = (12.0, 12.0)
    fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, sharex=True)

    font = {'color':  'black',
            'weight': 'normal',
            'size': 20,
            }
    tickfontsize=20
    ax0.errorbar(oligo['t'],
                 oligo['oligoB'],
                 yerr=oligo['oligoB_std'],
                 markersize=7,
                 fillstyle='none',
                 markeredgewidth=1,
                 fmt="ob")
    ax0.plot(oligo['t'],oligo['fitB'],"k",linewidth=2.0)
    ax0.tick_params(axis='y', labelsize=tickfontsize)
    ax0.locator_params(nbins=5, axis='y')
    ax0.set_ylabel("gB(t)",fontdict=font)
    ax1.errorbar(oligo['t'],
                 oligo['oligoR'],
                 yerr=oligo['oligoR_std'],
                 markersize=7,
                 fillstyle='none',
                 markeredgewidth=1,
                 fmt="or")
    ax1.plot(oligo['t'],oligo['fitR'],"k",linewidth=2.0)
    ax1.tick_params(axis='y', labelsize=tickfontsize)
    ax1.locator_params(nbins=5, axis='y')
    ax1.set_ylabel("gR(t)",fontdict=font)
    ax2.errorbar(oligo['t'],
                 oligo['oligoBR'],
                 yerr=oligo['oligoBR_std'],
                 markersize=7,
                 fillstyle='none',
                 markeredgewidth=1,
                 fmt="og")
    ax2.plot(oligo['t'],oligo['fitBR'],"k",linewidth=2.0)
    ax2.set_ylim(1.0,1.2)
    ax2.tick_params(axis='x', labelsize=tickfontsize)
    ax2.tick_params(axis='y', labelsize=tickfontsize)
    ax2.set_ylabel("gBR(t)",fontdict=font)
    ax2.set_xscale('log')
    ax2.set_xlabel('t in sec',fontdict=font)
    plt.tight_layout()
    plt.savefig(outputfile+'.png')
    plt.savefig(outputfile+'.pdf')
```

```python
makefig("oligo_fits_2_2.csv","oligoDA")
```
